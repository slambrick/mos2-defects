```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Fitted data for defect rate

The fitted data for diffraction intensity as a function defect rate is in the numpy binary files here. This script simply creates quick plots of that data.

```python
lower_120 = np.load("fit_data/data_fitted_defect_area_120-c_lowerbound.npy")
upper_120 = np.load("fit_data/data_fitted_defect_area_120-c_upperbound.npy")
c_120 = np.load("fit_data/data_fitted_defect_area_120-c.npy")
x = np.load("fit_data/data_fitted_xaxis.npy")

f, a = plt.subplots()
a.plot(x, c_120, label = "Average of fit")
a.plot(x, lower_120, '--', label = "Lower bound")
a.plot(x, upper_120, '--', label = "Upper bound")
a.set_xlabel('Defect densuity/$\\times 10^{14}\\,\\mathrm{cm}^{-2}$')
a.set_ylabel('Diffraction intensity/ arb.')
a.set_xlim(0, 2)
a.grid(True)
a.set_title("120C")
```
